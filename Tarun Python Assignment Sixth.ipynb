```python
#21. Write a function that takes a number and returns its factorial.
```

```python
def factorial(n):
    if n<0:
        return "Factorial not defined for negative number"
    if (n==0 or n==1):
        return 1
    return n*factorial(n-1)
factorial(5)
    
    
```

```python
#26. Write a function that takes a number and returns its reverse.
```

```python
def reverse_number(n):
    sign = -1 if n<0 else 1
    n_str=str(abs(n))
    reverse_str=n_str[::-1]
    return sign*int(reverse_str)
reverse_number(1234)
reverse_number(-1234)
```

```python
#34. Write a function that takes a number and returns both count of digits and sum of digits.
```

```python
def count_sum(num):
    count=0
    sum=0
    num=abs(num)
    while num>0:
        digit=num%10
        sum=sum+digit
        count=count+1
        num=num//10
    return sum,count
n=int(input("enter a number"))
count_sum(n)
```

```python
#35. Write a function that takes two numbers and returns their HCF using logic (not built-in).
```

```python
def find_hcf(a,b):
    hcf=1
    for i in range(1, min(a,b)+1):
        if a%i==0 and b%i==0:
            hcf=i
    return hcf
num1=(input("Enter first number"))
num2=(input("Enter second number"))
print(find_hcf(num1,num2))
```
